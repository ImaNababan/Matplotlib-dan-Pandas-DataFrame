```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(1, 11)

fig, ax1 = plt.subplots()

ax1.plot(x, np.exp(x), label='exp')
ax1.set_xlabel('Sumbu X')
ax1.set_ylabel('exp')

ax2 = ax1.twinx()
ax2.plot(x, np.log(x),'ro-', label='log')
ax2.set_ylabel('log')

fig.legend(loc='upper left')


plt.show()
```

```python
def GFG1(temp): 
	return (5. / 9.) * (temp - 32) 

def GFG2(ax1): 
	y1, y2 = ax1.get_ylim() 
	ax2.set_ylim(GFG1(y1), GFG1(y2)) 
	ax2.figure.canvas.draw() 

fig, ax1 = plt.subplots() 
ax2 = ax1.twinx() 

ax1.callbacks.connect("ylim_changed", GFG2) 
ax1.plot(np.linspace(0, 120, 100)) 
ax1.set_xlim(30, 100) 

ax1.set_ylabel('Fahrenheit') 
ax2.set_ylabel('Celsius') 

fig.suptitle('matplotlib.pyplot.twinx() function Example\n\n', fontweight ="bold") 
plt.show() 
```
